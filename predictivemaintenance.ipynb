```python
# delete unnecessary columns
df.drop(columns=['UDI', 'Product ID', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF'], inplace=True)
df.head()
```

```python
df["Type"].unique()
```

```python
Type_dict = {"L": 0, "M": 1, "H": 2}
df["Type"] = df["Type"].map(Type_dict)
```

```python
df
```

```python
df['Temp_Difference'] = df['Process temperature [K]'] - df['Air temperature [K]']

df['Power_Index'] = df['Rotational speed [rpm]'] * df['Torque [Nm]']

df['Torque_RPM_Ratio'] = df['Torque [Nm]'] / df['Rotational speed [rpm]']
```

```python
df['Machine failure'].value_counts()
```

```python
# Dependent Variable (Target)
y = df[['Machine failure']]

# Independent Variables (Features)
x = df.drop('Machine failure', axis=1)

print("Independent columns:", x.shape)
print("Dependent columns:", y.shape)
```

```python
print("Independent Variables:", x.columns)
print("Dependent variables:", y.columns)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
x_test.shape,y_test.shape,x_train.shape,y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Model Create
model = RandomForestClassifier(random_state=42)

# Train Model
model.fit(x_train, y_train)

# Prediction
y_pred = model.predict(x_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Accuracy (%):", accuracy * 100)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
import pandas as pd

importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': model.feature_importances_
})

importance = importance.sort_values('Importance', ascending=False)

print(importance)
```

```python
importance.plot(
    x='Feature',
    y='Importance',
    kind='bar',
    figsize=(10,5)
)
plt.title("Feature Importance")
plt.show()
```
